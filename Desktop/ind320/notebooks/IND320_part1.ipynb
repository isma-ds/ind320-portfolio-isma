## 3. Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

%matplotlib inline
```

## 4. Load Data
Place `open-meteo-subset.csv` next to your Streamlit app or provide a path here.

```python
csv_path = Path('../streamlit_app/open-meteo-subset.csv')  # Adjust the path if your file is located somewhere else.
df = pd.read_csv(csv_path)
df.head()
```

## 5. Quick Summary & Preview

```python
# Print the number of rows and columns in the dataset
print("Shape:", df.shape)

# Print the list of all column names
print("\nColumns:", list(df.columns))

# Display detailed summary statistics for all columns
display(df.describe(include="all"))
```

## 6. Plot Each Column Separately

```python
# Loop through each column in the dataframe
for col in df.columns:
    # Check if the column contains numeric data
    if pd.api.types.is_numeric_dtype(df[col]):
        # Create a new figure for each plot
        plt.figure()
        plt.plot(df[col])
        plt.title(col)
        plt.xlabel("Index")  # Label the x-axis
        # Label the y-axis
        plt.ylabel("Value")
        # Show the plot
        plt.show()
    else:
        # Skip columns that are not numeric
        print(f"Skipping non-numeric column: {col}")
```

## 7. Plot All Columns Together

```python
plt.figure()  # Create a new figure for the plot

for col in df.columns:  # Loop through each column in the DataFrame
    if pd.api.types.is_numeric_dtype(df[col]):  # Check if column is numeric
        plt.plot(df[col], label=col)  # Plot the numeric column with a label

plt.title("All Numeric Columns")  # Add a title to the plot
plt.xlabel("Index")  # Label the x-axis
plt.ylabel("Value")  # Label the y-axis
plt.legend()  # Show legend for all plotted columns
plt.show()  # Display the plot
```
